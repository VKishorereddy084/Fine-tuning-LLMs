```python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer, pipeline
import torch

model = "Kishorereddy123/llama2_finetuned"

tokenizer = AutoTokenizer.from_pretrained(model)

pipeline = pipeline(
    "text-generation", #task
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=150,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature':0})
```

```python
from langchain import PromptTemplate,  LLMChain

template = """
You are an helpful python AI. Help users with their queries.If you dont know the answer say I don't know.
Task: {question}
Answer:"""
prompt = PromptTemplate(template=template, input_variables=["question"])

llm_chain = LLMChain(prompt=prompt, llm=llm)

question = "Explain the differences between syntax errors, logic errors, and semantic errors in Python programming and provide examples for each type?"

print(llm_chain.run(question))
```
