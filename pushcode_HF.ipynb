```python
from datasets import load_dataset
import re

# Load the dataset
dataset = load_dataset('Kishorereddy123/accurate_QA')

# Shuffle the dataset and slice it
dataset = dataset['train'].shuffle(seed=42).select(range(86))

def transform_conversation(example):
    # Directly use the Question_Answer text
    conversation_text = example['Question_Answer']
    # Assume the format is "Question: <question> Answer: <answer>"
    question_text = conversation_text.split('Answer:')[0].replace('Question:', '').strip()
    answer_text = conversation_text.split('Answer:')[1].strip() if 'Answer:' in conversation_text else ''

    reformatted_text = f'<s>[INST] {question_text} [/INST] {answer_text} </s>'

    return {'Question_Answer': reformatted_text}

# Apply the transformation
transformed_dataset = dataset.map(transform_conversation)

print(transformed_dataset['train'][1])
```
